```python
df_1 = df[(df['Area']=='Afghanistan') &  (df['Element']=='Temperature change') &  (df['Months Code']<7013)]
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Data for plotting
t = df_1['Months']
s = df_1['Y1961']

fig, ax = plt.subplots()
ax.plot(t, s)
fig.autofmt_xdate()
ax.set(xlabel='Months', ylabel='Temperature change',
       title='(Afghanistan) Months vs Temperature change')
ax.grid()
fig.savefig("test.png")
plt.show()
```
